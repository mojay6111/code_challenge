
# üêç **Python ‚Äî Validate a FizzBuzz Sequence**

````python
def is_valid_fizzbuzz_sequence(arr):
    # 1. Check if array is empty or does not begin with 1
    if len(arr) == 0 or arr[0] != 1:
        return False

    # 2. Loop through expected numbers from 1 to len(arr)
    for i in range(1, len(arr) + 1):

        # 3. Determine what the correct FizzBuzz value should be
        if i % 15 == 0:
            expected = "FizzBuzz"
        elif i % 3 == 0:
            expected = "Fizz"
        elif i % 5 == 0:
            expected = "Buzz"
        else:
            expected = i

        # 4. Compare expected value with the actual array item
        if arr[i - 1] != expected:
            return False

    return True
````

````python
# Example usage:
print(is_valid_fizzbuzz_sequence([1, 2, "Fizz", 4, "Buzz", "Fizz", 7, 8, "Fizz", "Buzz", 11, "Fizz", 13, 14, "FizzBuzz"]))
# True

# Additional example usages:
print(is_valid_fizzbuzz_sequence([1, 2, "Fizz", 4, "Buzz", "Fizz", 7, 8, "Fizz", "Buzz", 11, "Fizz", 13, 14, "FizzBuzz", 16]))
# True

# Empty array
print(is_valid_fizzbuzz_sequence([]))
# False

# Incorrect FizzBuzz value
print(is_valid_fizzbuzz_sequence([1, 2, "Fizz", 4, "Buzz", "Fizz", 7, 8, "Fizz", "Buzz", 11, "Fizz", 13, 15, "FizzBuzz"]))
# False

# Extra value at the end
print(is_valid_fizzbuzz_sequence([1, 2, "Fizz", 4, "Buzz", "Fizz", 7, 8, "Fizz", "Buzz", 11, "Fizz", 13, 14, "FizzBuzz", "Buzz"]))
# False
````

# üß† **Detailed Explanation of Every Part**

---

## **1. Validate the start of the sequence**

```python
if len(arr) == 0 or arr[0] != 1:
    return False
```

FizzBuzz must start with the number **1**, so two checks are made:

* ‚ùå The array cannot be empty
* ‚ùå The first value must be `1`

If either fails ‚Üí the sequence is invalid immediately.

---

## **2. Loop through expected values**

```python
for i in range(1, len(arr) + 1):
```

If the array has length **N**, then the expected FizzBuzz sequence must include values for:

```
1, 2, 3, ... N
```

Since Python‚Äôs `range` excludes the end, we use `len(arr) + 1`.

---

## **3. Compute the correct FizzBuzz result**

```python
if i % 15 == 0:
    expected = "FizzBuzz"
elif i % 3 == 0:
    expected = "Fizz"
elif i % 5 == 0:
    expected = "Buzz"
else:
    expected = i
```

This part enforces FizzBuzz rules:

* Divisible by **15** ‚Üí `"FizzBuzz"`
  (we check this first to avoid double matching)
* Divisible by **3** ‚Üí `"Fizz"`
* Divisible by **5** ‚Üí `"Buzz"`
* Otherwise, it stays a number

Example:

* i = 3 ‚Üí `"Fizz"`
* i = 5 ‚Üí `"Buzz"`
* i = 15 ‚Üí `"FizzBuzz"`
* i = 7 ‚Üí `7`

---

## **4. Compare expected vs actual**

```python
if arr[i - 1] != expected:
    return False
```

Because array indices start at **0**, while our loop starts at **1**:

```
i = 1 ‚Üí arr[0]
i = 2 ‚Üí arr[1]
i = 3 ‚Üí arr[2]
```

If any value does **not** match what we expect ‚Üí invalid sequence ‚Üí `False`.

---

## **5. Everything matched**

```python
return True
```

If the loop finishes without mismatches, then the entire sequence is correct.
