# ✅ **1. FizzBuzz (Simple Explanation)**

You are given a number **n**.

You must create an array of values from **1 up to n**.
But while listing the numbers:

* If a number is a **multiple of 3**, replace it with **"Fizz"**.
* If a number is a **multiple of 5**, replace it with **"Buzz"**.
* If a number is a **multiple of BOTH 3 and 5** (i.e., 15, 30, 45...), replace it with **"FizzBuzz"**.
* If a number is not a multiple of 3 or 5, just keep the number.

This is the classic **FizzBuzz** problem, common in coding interviews.

Example for n = 15:

```
[
  1, 2, "Fizz", 4, "Buzz",
  "Fizz", 7, 8, "Fizz", "Buzz",
  11, "Fizz", 13, 14, "FizzBuzz"
]
```

---



# ✅ **2. JavaScript Code (JS)**

````javascript
function fizzBuzz(n) {
  const result = [];

  for (let i = 1; i <= n; i++) {
    if (i % 15 === 0) {
      result.push("FizzBuzz");
    } else if (i % 3 === 0) {
      result.push("Fizz");
    } else if (i % 5 === 0) {
      result.push("Buzz");
    } else {
      result.push(i);
    }
  }

  return result;
}
````

# ✅ **3. Explanation of Every Part of the Code**

### **`function fizzBuzz(n)`**

* Creates a function named **fizzBuzz** that accepts one input: the integer **n**.

---

### **`const result = [];`**

* Creates an empty array where we will store our final output.

---

### **`for (let i = 1; i <= n; i++)`**

* A loop that starts from **1** and counts up to **n**.
* `i` represents the current number we are examining.

---

### **`if (i % 15 === 0)`**

* `%` is the modulo operator → it gives the remainder.
* If a number is divisible by **both 3 and 5**, it must be divisible by **15**.
* So here, we check for **FizzBuzz first**.

Why first?
Because if we check 3 or 5 first, we will never reach this combined case.

---

### **`result.push("FizzBuzz");`**

* Adds the word `"FizzBuzz"` to the result array.

---

### **`else if (i % 3 === 0)`**

* If the number divides by **3** only → push `"Fizz"`.

---

### **`else if (i % 5 === 0)`**

* If the number divides by **5** only → push `"Buzz"`.

---

### **`else { result.push(i); }`**

* If none of the above apply, simply add the number.

---

### **`return result;`**

* Returns the full array.
