# âœ… **Problem Summary**

You are given:

* **message string** â†’ e.g. `"Hello World Again"`
* **validation string** â†’ e.g. `"HWA"`

A message is **valid** if:

1. Each word in the message starts with the **corresponding** character in the validation string.
2. **Order matters**.
3. **Case-insensitive**.
4. Words are separated by **single spaces**.

---

# âœ… **JavaScript Solution**

````javascript
function isValidMessage(message, validation) {
    // Step 1: split the message into individual words
    const words = message.split(" ");

    // Step 2: if number of words !== number of validation letters â†’ invalid
    if (words.length !== validation.length) {
        return false;
    }

    // Step 3: Compare each wordâ€™s first letter with the validation letter
    for (let i = 0; i < words.length; i++) {
        const wordFirstLetter = words[i][0].toLowerCase();
        const validLetter = validation[i].toLowerCase();

        if (wordFirstLetter !== validLetter) {
            return false;   // Mismatch â†’ invalid message
        }
    }

    // Step 4: All checks passed
    return true;
}
````

### **1. Split the message**

```js
const words = message.split(" ");
```

* Turns `"Hello World Again"`
  into
  `["Hello", "World", "Again"]`.

---

### **2. Validate count**

```js
if (words.length !== validation.length) return false;
```

* If message has 3 words, validation must have 3 letters.
* If not â†’ automatically invalid.

---

### **3. Compare each word**

```js
for (let i = 0; i < words.length; i++) {
    const wordFirstLetter = words[i][0].toLowerCase();
    const validLetter = validation[i].toLowerCase();
```


* Convert both letters to lowercase (case-insensitive).
* Compare the first letter of each message word with the matching validation letter.

---

### **4. If all letters match**

Return `true`.
